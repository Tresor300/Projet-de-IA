```python
import pandas as pd

# Chargement du jeu de données brut
df_brut = pd.read_csv('export_IA.csv')

# Affichage de toutes les colonnes disponibles
print(df_brut.columns.tolist())
```

```python
# Sélection des colonnes exactes trouvées dans votre fichier
colonnes_utiles = [
    'consolidated_latitude', 
    'consolidated_longitude', 
    'implantation_station', 
    'puissance_nominale'
]

# Création du DataFrame dédié au Besoin 1
df_besoin1 = df_brut[colonnes_utiles].copy()

# Affichage des premières lignes pour vérifier que l'extraction a fonctionné
print("--- Aperçu du nouveau tableau filtré ---")
print(df_besoin1.head())

# Vérification critique : y a-t-il des données manquantes dans ces 4 colonnes ?
print("\n--- Décompte des valeurs nulles (NaN) ---")
print(df_besoin1.isnull().sum())
```

```python
import plotly.express as px
import plotly.io as pio

# LA SOLUTION : On force Plotly à contourner VS Code et à utiliser le navigateur web
pio.renderers.default = "browser"

# 1. Graphique de la répartition des types d'implantation
fig_impl = px.histogram(
    df_besoin1, 
    x='implantation_station', 
    color='implantation_station',
    title="Répartition des bornes par type d'implantation",
    labels={'implantation_station': "Type d'implantation"},
    text_auto=True
)
fig_impl.update_layout(xaxis_categoryorder='total descending')
fig_impl.show() # Cela va ouvrir un nouvel onglet dans votre navigateur

# 2. Graphique de la distribution des puissances nominales
fig_puiss = px.histogram(
    df_besoin1, 
    x='puissance_nominale', 
    title="Distribution de la puissance nominale (en kW)",
    nbins=50,
    labels={'puissance_nominale': 'Puissance Nominale (kW)'}
)
fig_puiss.show() # Cela va ouvrir un deuxième onglet
```

```python
import folium
from folium.plugins import HeatMap, MarkerCluster

# 1. Calcul du centre de la carte
center_lat = df_besoin1['consolidated_latitude'].mean()
center_lon = df_besoin1['consolidated_longitude'].mean()

# ==========================================
# CARTE 1 : LES TYPES D'IMPLANTATION
# ==========================================
print("Génération de la carte des implantations...")
carte_types = folium.Map(location=[center_lat, center_lon], zoom_start=6)

# On liste tous les types d'implantation uniques
types_uniques = df_besoin1['implantation_station'].unique()

# On crée un groupe/cluster pour chaque type
for type_impl in types_uniques:
    df_filtre = df_besoin1[df_besoin1['implantation_station'] == type_impl]
    
    # Création du cluster (le regroupement automatique demandé par le client)
    cluster = MarkerCluster(name=str(type_impl))
    
    # On ajoute les points de ce type dans le cluster
    for _, row in df_filtre.iterrows():
        folium.CircleMarker(
            location=[row['consolidated_latitude'], row['consolidated_longitude']],
            radius=4,
            color='blue', # La couleur importe peu, le cluster regroupe par type
            fill=True,
            tooltip=f"Puissance : {row['puissance_nominale']} kW"
        ).add_to(cluster)
        
    # On ajoute le cluster terminé à la carte
    cluster.add_to(carte_types)

# Ajout du menu en haut à droite pour filtrer les types
folium.LayerControl().add_to(carte_types)

# Sauvegarde
carte_types.save('carte_implantations.html')


# ==========================================
# CARTE 2 : LA HEATMAP (CARTE DE CHALEUR)
# ==========================================
print("Génération de la carte de chaleur...")
carte_chaleur = folium.Map(location=[center_lat, center_lon], zoom_start=6)

heat_data = df_besoin1[['consolidated_latitude', 'consolidated_longitude']].values.tolist()
HeatMap(heat_data, radius=15, blur=10).add_to(carte_chaleur)

# Sauvegarde
carte_chaleur.save('carte_chaleur.html')

print("Terminé ! 'carte_implantations.html' et 'carte_chaleur.html' ont été générés.")
```
